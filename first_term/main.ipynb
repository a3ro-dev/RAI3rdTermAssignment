```python
import matplotlib.pyplot as plt
import numpy as np  
import pandas as pd
import seaborn as sns
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
```

```python
data = {"Name":["a", "b", "c", "d"], "Marks":[85,70,90,60]}
df = pd.DataFrame(data)
df.Marks.plot(kind='bar')
```

```python
tips = sns.load_dataset("tips")
sns.barplot(x="day", y="total_bill", data=tips)
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 4, 5, 6]
y2 = [3, 4, 5, 6, 7]
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Lines Plot')
plt.legend()
plt.show()
```

```python
sizes = [40, 30, 20, 10]
labels = ["A","B","C","D"]
explode = (0.1, 0, 0, 0)  # highlight first slice
plt.pie(sizes, labels=labels, explode=explode, autopct="%1.1f%%")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fmri = sns.load_dataset("fmri")
sns.lineplot(x="timepoint", y="signal", data=fmri)
plt.show()
```

```python
df = pd.DataFrame({"Age":[18,20,22,19,21,20,23,22,19]})
df["Age"].plot(kind="hist", bins=5, edgecolor="black")
plt.show()
```

```python
import seaborn as sns
iris = sns.load_dataset("iris")
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=iris)
plt.show()
```

```python
data = {"Day": pd.date_range(start="2025-01-01", periods=7, freq="D"),
        "Visitors": [120,135,150,160,155,170,180]}
df = pd.DataFrame(data)

df.plot(x="Day", y="Visitors", marker="o", title="Website Visitors Over Time")
plt.show()
```
