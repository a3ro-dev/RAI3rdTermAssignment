1. Write a Python program that slices a list of email addresses to extract `usernames` and `domains`. Then, analyze and visualize the frequency of email domains using the following plots with `matplotlib` and `seaborn`:<br>
a. **Bar Plot**<br>
&nbsp;&nbsp;&nbsp;&nbsp;X-axis: `Domain names` (e.g., gmail.com, yahoo.com)<br>
&nbsp;&nbsp;&nbsp;&nbsp;Y-axis: `Frequency` (number of times each domain appears)<br>
b. **Line Plot**<br>
&nbsp;&nbsp;&nbsp;&nbsp;X-axis: `Domain names`<br>
&nbsp;&nbsp;&nbsp;&nbsp;Y-axis: `Frequency`<br>
c. **Scatter Plot**<br>
&nbsp;&nbsp;&nbsp;&nbsp;X-axis: `Encoded domain index` (e.g., 0 for gmail.com, ...)<br>
&nbsp;&nbsp;&nbsp;&nbsp;Y-axis: `Frequency`<br>
d. **Pie Chart**<br>
&nbsp;&nbsp;&nbsp;&nbsp;Slices: `Domain names`<br>
&nbsp;&nbsp;&nbsp;&nbsp;Size of each slice: `Frequency`<br>
e. **Box Plot**<br>
&nbsp;&nbsp;&nbsp;&nbsp;X-axis: `Domain`<br>
&nbsp;&nbsp;&nbsp;&nbsp;Y-axis: `Frequency of emails`<br>
f. **Histogram**<br>
&nbsp;&nbsp;&nbsp;&nbsp;X-axis: `Frequency Values`<br>
&nbsp;&nbsp;&nbsp;&nbsp;Y-axis: `Number of domains with those frequencies`<br>


```python
emails = [
    "alice@gmail.com","bob@yahoo.com","carol@gmail.com","dave@outlook.com",
    "erin@yahoo.com","frank@gmail.com","grace@proton.me","heidi@outlook.com",
    "ivan@gmail.com","judy@yahoo.com","mallory@icloud.com","oscar@gmail.com",
    "peggy@duck.com","trent@yahoo.com","victor@proton.me","walter@gmail.com"
]
usernames = [e.split('@')[0] for e in emails]
domains = [e.split('@')[1] for e in emails]
counts = pd.Series(domains).value_counts().sort_index()
idx = np.arange(len(counts))

fig, axes = plt.subplots(2, 3, figsize=(10.5, 6.5))
ax = axes[0,0]
sns.barplot(x=counts.index, y=counts.values, color='C0', ax=ax)
ax.set_xticks(np.arange(len(counts.index)))
ax.set_xticklabels(list(counts.index), rotation=45, ha='right')
ax.set_xlabel('Domain names'); ax.set_ylabel('Frequency'); ax.set_title('Bar Plot')

ax = axes[0,1]
ax.plot(idx, counts.values, marker='o')
ax.set_xticks(idx)
ax.set_xticklabels(list(counts.index), rotation=45, ha='right')
ax.set_xlabel('Domain names'); ax.set_ylabel('Frequency'); ax.set_title('Line Plot')

ax = axes[0,2]
ax.scatter(idx, counts.values)
ax.set_xlabel('Encoded domain index'); ax.set_ylabel('Frequency'); ax.set_title('Scatter Plot')

ax = axes[1,0]
ax.pie(counts.values, labels=counts.index, autopct='%1.0f%%')
ax.set_title('Pie Chart')

ax = axes[1,1]
sns.boxplot(x=counts.index, y=counts.values, ax=ax)
ax.set_xticks(np.arange(len(counts.index)))
ax.set_xticklabels(list(counts.index), rotation=45, ha='right')
ax.set_xlabel('Domain'); ax.set_ylabel('Frequency of emails'); ax.set_title('Box Plot')

ax = axes[1,2]
ax.hist(counts.values, bins=range(1, counts.max()+2))
ax.set_xlabel('Frequency Values'); ax.set_ylabel('Number of domains'); ax.set_title('Histogram')

plt.tight_layout(pad=0.6, w_pad=0.6, h_pad=0.8)
```

3. Write a Python program to analyze climate using `matplotlib` and `seaborn`.<br>

&nbsp;&nbsp;&nbsp;&nbsp;Visualize rainfall for each month.<br>

4. Write a Python program to create a **Rock, Paper, Scissor** game and visualize the results using `matplotlib`.<br>
Requirements:<br>
&nbsp;&nbsp;&nbsp;&nbsp;- Simulate multiple rounds of the game between a user and computer.<br>
&nbsp;&nbsp;&nbsp;&nbsp;- For each round, randomly select moves for both user and computer from `['Rock', 'Paper', 'Scissor']`.<br>
&nbsp;&nbsp;&nbsp;&nbsp;- Determine the winner for each round.<br>
&nbsp;&nbsp;&nbsp;&nbsp;- Count the number of wins for `User`, `Computer`, and `Draw`.<br>
&nbsp;&nbsp;&nbsp;&nbsp;- Visualize the results using a **bar plot** with `matplotlib` and `seaborn`.<br>

```python
choices = ['Rock','Paper','Scissor']
rounds = 20
user_moves = [random.choice(choices) for _ in range(rounds)]
comp_moves = [random.choice(choices) for _ in range(rounds)]
res = []
for u,c in zip(user_moves, comp_moves):
    if u == c:
        res.append('Draw')
    elif (u=='Rock' and c=='Scissor') or (u=='Paper' and c=='Rock') or (u=='Scissor' and c=='Paper'):
        res.append('User')
    else:
        res.append('Computer')
summary = pd.Series(res).value_counts().reindex(['User','Computer','Draw']).fillna(0)
plt.figure(); sns.barplot(x=summary.index, y=summary.values, color='C0'); plt.xlabel('Result'); plt.ylabel('Count'); plt.title('Rock Paper Scissor Results'); plt.tight_layout()
```
